```python
import pandas as pd

df = pd.DataFrame.from_dict(response_data).T.reset_index()
df = df.rename(columns={'index':'name'})
df.head()
```

```python
# separate positive, negative, medical:
positive = []
negative = []
medical = []

for index, row in df.iterrows():
    positive.append(row['effects']['positive'])
    negative.append(row['effects']['negative'])
    medical.append(row['effects']['medical'])

df['positive_effects'] = positive
df['negative_effects'] = negative
df['medical_uses'] = medical

df.head()
```

```python
# remove 'effects' column:
df = df.drop('effects', axis=1)
```

```python
strain_df = df.copy()
print(strain_df.shape)
strain_df.head()
```

```python
# kaggle dataset:
kaggle_df = pd.read_csv('./cannabis.csv')
kaggle_df.head()
```

```python
# clean up column names:
kaggle_df = kaggle_df.rename(columns={'Strain':'name', 'Type':'race'})
```

```python
# fix format of names so can be merged:
kaggle_df['name'] = kaggle_df['name'].apply(lambda x: x.strip().replace('-',' '))
```

```python
# merge df's:
merged = strain_df.merge(kaggle_df, how='inner', on=['name', 'race'])

print(merged.shape)
merged.head(3)
```

```python
# combine flavors:
for index, row in merged.iterrows():
    for flav in row['Flavor'].split(','):
        if flav not in row['flavors']:
            row['flavors'].append(flav)

merged = merged.drop('Flavor', axis=1)
```

```python
# combine positive effects:
for index, row in merged.iterrows():
    for effect in row['Effects'].split(','):
        if effect not in row['positive_effects']:
            row['positive_effects'].append(effect)
            
merged = merged.drop('Effects', axis=1)
```

```python
merged = merged.set_index('id')
```
